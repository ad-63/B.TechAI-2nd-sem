# a. Commit():
commit() is a function which is used t save the transactions permanently in the database after performing any updates or making any changes.

-> it moves the pointer to the next tranaction state affter adding commit().

syntax:
connection.commiit()

# b.rollback():
there is high chances of failure or error while writing the program and if any changes made is not fully completely then those transaction should be rollbacked.

-> rollback() is a function which used to rollbacl any transaction if it is not fully completed to it just previous state.
-> incase if the transaction is committed then it cannot be rollbacked to the previous state.

syntax:
connection.rollback()

-> most of the time we do the rollback operation in the exception section because rollback only happens when any failure is seen. if everything is perfectly then it would be committed not rollbacked.

difference between comit ad rollback
# commit()
i.commit() is a function which saves the transaction permanently to the database.
ii. if everything is worked perfectly then we do the commit() to save the changed.
iii. moves the transaction to the next transaction state i.e: t=t+1.

# rollback()
1. rollback() is a fucntion which does not save the trancastion rather moves to the previous latest state.
2. if anything failes while performing any operation then we do the rollback inorder to let it go to the previous state.
3. rollback happens only when there is failure or any exception.
