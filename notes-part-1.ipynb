```python
# to store details of two cars and display the highest priced car.
class car:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def display(self):
        # Changed self.name to self.brand
        print(f"brand={self.brand}\nprice={self.price}")
         
c1 = car("mercedes", 1000000)
c2 = car("hyundai", 800000)

print(c1.price)
print(c2.price)

if c1.price > c2.price:
    c1.display()
else:
    c2.display()
```

```python
# wap to input N students detail and display the highest age student details.

class student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name={self.name}\nAge={self.age}")

size = int(input("enter how many students to store: "))
student_list = []

for i in range(size):
    name = input(f"enter {i+1} student name: ")
    age = int(input(f"enter {i+1} student age: "))
    student_list.append(student(name, age))

# declaring highest age for comparison 
highest_age = student_list[0].age
for std in student_list:
    if std.age > highest_age:
        highest_age = std.age

# printing highest age student details
for std in student_list:
    if highest_age == std.age:
        std.display()
```

# Type casting is the process of converting one data type to another.
### types of type casting:
# implicit type casting: This is done automatically by the Python interpreter when it encounters an operation that involves different data types.
### For example, when you add an integer and a float, the integer is automatically converted to a float before the operation is performed.


```python
num1 = 5  # integer
num2 = 3.14  # float
sum = num1 + num2  # implicit type casting: num1 is converted to float before addition
print("The sum of", num1, "and", num2, "is", sum)
#explicit type casting: This is done manually by the programmer using built-in functions like int(), float
num1 = 5  # integer
num2 = 3.14  # float            
sum = int(num1) + int(num2)  # explicit type casting: both num1 and num2 are converted to integers before addition
print("The sum of", num1, "and", num2, "is", sum)
```

# list slicing:
 List slicing is a technique used to extract a portion of a list by specifying a range of indices. The syntax for list slicing is: list[start:stop:step]


```python
# WAP to add and delete elements from an list.
my_list = [1, 2, 3, 'a', 'b', 'c']
print(my_list[0:5:2]) # This will return every second element from index 0 to 4 (1, 3, 'b')
print("Original list:", my_list)
# Adding an element to the list
my_list.append('d')
print("List after adding an element:", my_list)
# Deleting an element from the list
my_list.remove(2)   
print("List after deleting an element:", my_list)  
```

# Tuple:
 A tuple is similar to a list but is immutable, meaning it cannot be changed after creation. Tuples are defined using parentheses ().
### eg: my_tuple = (1, 2, 3, 'a', 'b', 'c')
### eg: students = ('Alice', 'Bob', 'Charlie')
### eg:subjects = ('Math', 'Science', 'History')

```python
#WAP to add and delete elements from a tuple.
my_tuple = (1, 2, 3, 'a', 'b', 'c')
print("Original tuple:", my_tuple)
print(id(my_tuple))  # This will print the memory address of the original tuple
# Adding an element to the tuple (this will create a new tuple)     
my_tuple = my_tuple + ('d',)  # Note the comma to create a single-element tuple
print("Tuple after adding an element:", my_tuple)
print(id(my_tuple))  # This will print the memory address of the new tuple, which is different from the original tuple
# Deleting an element from the tuple (this will create a new tuple without the element)
my_tuple = tuple(x for x in my_tuple if x != 2)  # This will create a new tuple without the element '2'
print("Tuple after deleting an element:", my_tuple) 
#Proof that tuples are immutable by trying to change an element (this will raise an error)
try:
    my_tuple[0] = 10  # This will raise a TypeError because tuples are immutable
except TypeError as e:
    print("Error:", e)
```

### eg: my_set = {1, 2, 3, 'a', 'b', 'c'}
### eg: students = {'Alice', 'Bob', 'Charlie'}     
### eg:subjects = {'Math', 'Science', 'History'}



```python
#WAP to add and delete elements from a set.
my_set = {1, 2, 3, 'a', 'b', 'c'}
print("Original set:", my_set)
# Adding an element to the set
my_set.add('d')
print("Set after adding an element:", my_set)
# Deleting an element from the set
my_set.remove(2)
print("Set after deleting an element:", my_set)
```

# Dictionary:
 A dictionary is a collection of key-value pairs. It is unordered and mutable, meaning you can change its contents after creation. Dictionaries are defined using curly braces {} with key-value pairs separated by a colon :.


```python
#eg: my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
#eg: student_info = {'Alice': 20, 'Bob': 22, 'Charlie': 19}
#eg: subject_info = {'Math': 'A', 'Science': 'B', 'History': 'C'}
#WAP to add and delete elements from a dictionary.
my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
print("Original dictionary:", my_dict)
# Adding a key-value pair to the dictionary
my_dict['country'] = 'USA'      
print("Dictionary after adding a key-value pair:", my_dict)
# Deleting a key-value pair from the dictionary
del my_dict['age']
print("Dictionary after deleting a key-value pair:", my_dict)
```

```python
# WAP to make a atm system and exit the while loop when user presses 4 
"""1)whithdraw
2)check balance 
3)deposite 
4)exit """
balance = 1000.0  # Starting balance

while True:
    print("\n--- ATM System ---")
    print("1) Withdraw")
    print("2) Check Balance")
    print("3) Deposit")
    print("4) Exit")
    
    choice = input("Select an option (1-4): ")

    if choice == '1':
        amount = float(input("Enter amount to withdraw: "))
        if amount > balance:
            print("Insufficient funds!")
        else:
            balance -= amount
            print(f"${amount} withdrawn successfully.")

    elif choice == '2':
        print(f"Your current balance is: ${balance}")

    elif choice == '3':
        amount = float(input("Enter amount to deposit: "))
        balance += amount
        print(f"${amount} deposited successfully.")

    elif choice == '4':
        print("Thank you for using the ATM. Goodbye!")
        break  # This exits the while loop

    else:
        print("Invalid choice. Please try again.")

```

## multiple inheritance:
if we inherit the feature and methods from multiple parent class to any one child class then it is called multiple inheritance.


```python
class father:
    def land(self):
        print("land is the property of father")

class mother:
    def money(self):
        print("money is the property of mother")

class son(father,mother):
    pass

son=son()
son.land()  
son.money()       
```

```python
class person:
    def display(self,name):
        print("name=",name)

class student(person):
    pass

class teacher(person):
    pass
std=student()
std.display("ram")
teacher=teacher() 
teacher.display("hari")
```

## polymorphism: 
it is a feature of oop in which a "methord" have different forms.
it can be achived in two ways:
1. method overloading: when we have more than one method with the same name but different parameters in the same class then it is called method overloading.
2. method overriding: when we have a method in the parent class and the same method in the child class then it is called method overriding.


### method overloading:

```python
class animal:
     def sound(self):
         print("animal sound")

class dog(animal):
       def sound(self):
          print("dog barks")

class cat(animal):
        def sound(self):
            print("cat meows")     

dog=dog()
dog.sound()
cat=cat()
cat.sound()
```
