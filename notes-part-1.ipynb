```python
# to store details of two cars and display the highest priced car.
class car:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def display(self):
        # Changed self.name to self.brand
        print(f"brand={self.brand}\nprice={self.price}")
         
c1 = car("mercedes", 1000000)
c2 = car("hyundai", 800000)

print(c1.price)
print(c2.price)

if c1.price > c2.price:
    c1.display()
else:
    c2.display()
```

```python
# wap to input N students detail and display the highest age student details.

class student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name={self.name}\nAge={self.age}")

size = int(input("enter how many students to store: "))
student_list = []

for i in range(size):
    name = input(f"enter {i+1} student name: ")
    age = int(input(f"enter {i+1} student age: "))
    student_list.append(student(name, age))

# declaring highest age for comparison 
highest_age = student_list[0].age
for std in student_list:
    if std.age > highest_age:
        highest_age = std.age

# printing highest age student details
for std in student_list:
    if highest_age == std.age:
        std.display()
```

# Type casting is the process of converting one data type to another.
### types of type casting:
# implicit type casting: This is done automatically by the Python interpreter when it encounters an operation that involves different data types.
### For example, when you add an integer and a float, the integer is automatically converted to a float before the operation is performed.


```python
num1 = 5  # integer
num2 = 3.14  # float
sum = num1 + num2  # implicit type casting: num1 is converted to float before addition
print("The sum of", num1, "and", num2, "is", sum)
#explicit type casting: This is done manually by the programmer using built-in functions like int(), float
num1 = 5  # integer
num2 = 3.14  # float            
sum = int(num1) + int(num2)  # explicit type casting: both num1 and num2 are converted to integers before addition
print("The sum of", num1, "and", num2, "is", sum)
```

# Tuple:
 A tuple is similar to a list but is immutable, meaning it cannot be changed after creation. Tuples are defined using parentheses ().
### eg: my_tuple = (1, 2, 3, 'a', 'b', 'c')
### eg: students = ('Alice', 'Bob', 'Charlie')
### eg:subjects = ('Math', 'Science', 'History')

```python
#WAP to add and delete elements from a tuple.
my_tuple = (1, 2, 3, 'a', 'b', 'c')
print("Original tuple:", my_tuple)
print(id(my_tuple))  # This will print the memory address of the original tuple
# Adding an element to the tuple (this will create a new tuple)     
my_tuple = my_tuple + ('d',)  # Note the comma to create a single-element tuple
print("Tuple after adding an element:", my_tuple)
print(id(my_tuple))  # This will print the memory address of the new tuple, which is different from the original tuple
# Deleting an element from the tuple (this will create a new tuple without the element)
my_tuple = tuple(x for x in my_tuple if x != 2)  # This will create a new tuple without the element '2'
print("Tuple after deleting an element:", my_tuple) 
#Proof that tuples are immutable by trying to change an element (this will raise an error)
try:
    my_tuple[0] = 10  # This will raise a TypeError because tuples are immutable
except TypeError as e:
    print("Error:", e)
```

## THE EXACT SAME EXAMPLE (MULTIPLYING A NUMBER BY 2) IN ALL FOUR FUNCTION TYPES





```python
# ------------------------------------------------------------------------------
# 1) WITHOUT ARGUMENT AND WITHOUT RETURN TYPE
# ------------------------------------------------------------------------------
# Everything happens inside the function. 
# It picks its own number, does the math, and prints it.
def double_number_type_1():
    num = 5                      # Hardcoded inside the function
    result = num * 2             # Calculation
    print(f"[Type 1] Double of {num} is {result}")
    # No return statement


# ------------------------------------------------------------------------------
# 2) WITHOUT ARGUMENT AND WITH RETURN TYPE
# ------------------------------------------------------------------------------
# The function picks its own number and does the math, 
# but it sends the final answer back to the main program.
def double_number_type_2():
    num = 5                      # Hardcoded inside the function
    result = num * 2             # Calculation
    return result                # Sends the answer back


# ------------------------------------------------------------------------------
# 3) WITH ARGUMENT AND WITHOUT RETURN TYPE
# ------------------------------------------------------------------------------
# The main program provides the number (input). 
# The function does the math and prints it right there.
def double_number_type_3(num):
    result = num * 2             # Calculation using the input argument
    print(f"[Type 3] Double of {num} is {result}")
    # No return statement


# ------------------------------------------------------------------------------
# 4) WITH ARGUMENT AND WITH RETURN TYPE
# ------------------------------------------------------------------------------
# The main program provides the number (input).
# The function does the math and sends the answer back (output).
def double_number_type_4(num):
    result = num * 2             # Calculation using the input argument
    return result                # Sends the answer back




# ==============================================================================
# MAIN PROGRAM EXECUTION (How we call each type)
# ==============================================================================
if __name__ == "__main__":
    
    print("--- Executing Type 1 ---")
    # We pass nothing, and we get nothing back. It prints on its own.
    double_number_type_1()
    
    print("\n--- Executing Type 2 ---")
    # We pass nothing, but we must catch the returned answer in a variable.
    answer_from_2 = double_number_type_2()
    print(f"[Type 2] Main program received the answer: {answer_from_2}")
    
    print("\n--- Executing Type 3 ---")
    # We provide the number 5, but it prints the result on its own.
    double_number_type_3(5)
    
    print("\n--- Executing Type 4 ---")
    # We provide the number 5, AND we catch the returned answer in a variable.
    answer_from_4 = double_number_type_4(5)
    print(f"[Type 4] Main program received the answer: {answer_from_4}")
```

A process of inheriting the feature of parent class to child class is known as inherentance.

-> Inheritance removes the code duplicacy
-> Syntax:
                       class Person:

### Types of Inheritence :
A) Simple Inheritance :
                 when we inherit Feature of one class to another only one class then it is called simple inheritance :
                 Example :       [Parent Class]
                                       |
                                       |
                                  [child class]


B) Multilevel Inheritance :
           Deriving feature from one parent class which is deriving the feature from another it's parent class:

```python
# Example of Multi level Inheritance 
class Grandfather:
  def display(self):
    print("my land")
class Father(Grandfather):
  pass
class Son(Father):
  pass
son = Son()
son.display()
```

C.Multiple inheritance:  If we inherit the feature and methods from multiple parent class to any one child class then it is called multiple inheritance.

```python
class Father:
    def land(self):
        print("Land")
class Mother:
    def jewellery(self):
        print("Jewelleries")
class Daughter(Father,Mother):
    pass
daughter=Daughter()
daughter.land()
daughter.jewellery()
```

D.Derived iniheritance:  When one parent class is usedto inherit its features to two or more child classes then it is called derived inheritance .


```python
class Person: 
    def display(self,name):
        print("Name=",name)

class Student(Person):
    pass
class Teacher(Person):
    pass
std=Student()
 
std.display("Brishab")
teacher=Teacher()
teacher.display("Sunil")
```

 A process of binding data and methods into a single unit which protects the direct data access is called encapsulation.
 The variabes in python are public, protected and private.
 Access Modifires:-
                                                Access
 Modifiers   Symbols            Same class    Child class   Outside class
 public      variableName           T             T             T
 protected  _variableName           T             T             F
 private    __variableName          T             F             F

```python
class Father:
    def __init__(self):
        self.name="Brishab"
        self.__age=21
    def Set_age(self,age):
        self.age=age
father=Father()
father.__age=22
print(father.__age)
print(father._Father__age)
```

```python
class Father:
    def __init__(self):
        self.name = "Brishab"
        self.__age = 21  # This gets mangled to _Father__age
        
    def Set_age(self, age):
        self.__age = age  # Correctly updates the mangled variable
        
    def get_age(self):
        return self.__age  # FIXED: added the double underscores

father = Father()

# 1. Correct way to call the setter method
father.Set_age(22)

# 2. Using the getter method (The right way to access private data)
print("Age via getter:", father.get_age())  # Output: 22

# 3. Trying to access the raw double underscore name directly will still fail:
# print(father.__age) <-- This would throw an AttributeError

# 4. Accessing via the "Mangled" name (Sneaky, but works in Python)
print("Age via mangled name:", father._Father__age)  # Output: 22
```

```python
class Father:
    def __init__(self):
        self.name = "Brishab"
        self.__age = 20         # Private attribute (Mangled to _Father__age)
        self._gender = "Male"   # Protected attribute (Single underscore)

    def set_age(self, age):
        self.__age = age

    def get_age(self):
        return self.__age       # Fixed typo: added __

class Son(Father):
    def display(self):
        # FIX: Call self.get_age() instead of self.__age
        # FIX: Use self._gender to match how it was defined
        print(f"Name={self.name}\nAge={self.get_age()}\nGender={self._gender}")

# Execution
son = Son()
son.display()
```
